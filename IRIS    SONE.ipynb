```python
numerical = ["SepalLength", "SepalWidth","PetalLength","PetalWidth"]
numerical_axis = ["length of sepal", "width of sepal","length of petal","width of petal"]
```

```python
list(zip(numerical, numerical_axis))
```

```python
title_font = {"family":"calibri light","color":"darkblue","weight":"Bold","size":10}
axis_font = {"family":"calibri","color":"red","weight":"Bold","size":10}
```

```python
for i,z in list(zip(numerical,numerical_axis)):
  plt.figure(figsize=(10,10),dpi=90)
  sns.distplot(iris[i],hist_kws = dict(linewidth=1,edgecolor="r"))
  plt.title(i,fontdict = title_font)
  plt.xlabel(z,fontdict = axis_font)
  plt.ylabel("Density",fontdict = axis_font)
  plt.tight_layout()
  plt.show()
```
